```python
import pandas as pd
import re 
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
from nltk.stem import WordNetLemmatizer
from gensim.models import Word2Vec
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from nltk.tokenize import word_tokenize
import numpy as np
from tqdm import tqdm
```

```python
df = pd.read_csv("all_kindle_review .csv")
```

```python
df = df[["reviewText","rating"]]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["rating"].value_counts()
```

```python
df["rating"].unique()
```

```python
df['rating'] = df["rating"].apply(lambda x:1 if x>3 else 0)
# 1 if rating is > 3 , 0 for rating <= 3
```

```python
df.head()
```

```python
df['rating'].value_counts()
```

```python
df["reviewText"] = df["reviewText"].str.lower()
```

```python
#removing special character
df['reviewText'] = df["reviewText"].apply(lambda x: re.sub('[^a-z A-Z 0-9]+'," ",x))
df['reviewText'] = df["reviewText"].apply(lambda x: " ".join([y for y in x.split() if y not in stopwords.words('english')]))
df['reviewText'] = df["reviewText"].apply(lambda x: BeautifulSoup(x,"html.parser").get_text())
df["reviewText"] = df["reviewText"].apply(lambda x: ' '.join(x.split())) #remove additional spaces..
```

```python
df.head()
```

```python
wnl = WordNetLemmatizer()
```

```python
def lemmatize_words(text):
    return " ".join([wnl.lemmatize(word) for word in text.split()])
```

```python
df["reviewText"] = df["reviewText"].apply(lambda x:lemmatize_words(x))
```

```python
df.head()
```

```python
X = df["reviewText"]
y = df["rating"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20)
```

```python
#converting into vectors using BOW
bow = CountVectorizer()
X_train_bow = bow.fit_transform(X_train).toarray()
X_test_bow = bow.transform(X_test).toarray()
```

```python
tfidf = TfidfVectorizer()
X_train_tfidf = tfidf.fit_transform(X_train).toarray()
X_test_tfidf = tfidf.transform(X_test).toarray()
```

```python
mnb_bow = MultinomialNB().fit(X_train_bow,y_train)
```

```python
y_pred_bow = mnb_bow.predict(X_test_bow)
print(accuracy_score(y_pred_bow,y_test))
print(confusion_matrix(y_pred_bow,y_test))
print(classification_report(y_pred_bow,y_test))
```

```python
mnb_tfidf = MultinomialNB().fit(X_train_tfidf,y_train)
y_pred_tfidf = mnb_tfidf.predict(X_test_tfidf)
print(accuracy_score(y_pred_tfidf,y_test))
print(confusion_matrix(y_pred_tfidf,y_test))
print(classification_report(y_pred_tfidf,y_test))
```

```python
words = []
for sent in X:
    words.append(word_tokenize(sent))
```

```python
words
```

```python
w2v_model = Word2Vec(words,epochs=10)
```

```python
words[0]
```

```python
def avgword2vec(doc):
    # Filter words that are in the model's vocabulary
    vectors = [w2v_model.wv[word] for word in doc if word in w2v_model.wv.index_to_key]
    if len(vectors) > 0:
        return np.mean(vectors, axis=0)  # Compute mean along the correct axis
    else:
        return np.zeros(w2v_model.vector_size)  # Return a zero vector if no words are found

# Assuming `words` is a list of tokenized documents
X_w2v = []
for i in tqdm(range(len(words))):
    X_w2v.append(avgword2vec(words[i]))

# Convert X to a numpy array for further processing
X_w2v = np.array(X_w2v)
```
